# Data collection and preprocessing from the ChEMBL database - collected the bioatives of the target serotonin 5-HT2A receptor
p.s. to data scientists here the target is not as in ML, it A biological molecule (usually a protein) inside the body that a drug needs to bind to to have an effect.(The lock on a door).

```python
# Import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
target = new_client.target
target_query = target.search('5-HT2A')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
# Select only the human 5-HT2A receptor
human_5ht2a = targets[
    (targets['organism'] == 'Homo sapiens') &
    (targets['pref_name'] == '5-hydroxytryptamine receptor 2A')
]
```

```python
human_5ht2a
```

 ***As we see here we got 1 row after filtering to only human organism and the human receptor 5-hydroxytryptamine receptor 2A also it's already a single protein so we don't have to filter for it, now we'll look for its ChEMBLID to look for the right bioactives***

```python
chembl_id = human_5ht2a['target_chembl_id'].values[0]
print(f"human_5ht2a ChEMBL ID: {chembl_id}")
```

Let's look for the bioactives


```python
activity = new_client.activity
activity
```

```python
act_query = activity.search('CHEMBL224')
activities = pd.DataFrame.from_dict(act_query)
activities
```

```python
activities.columns
```

```python
df = activities[activities["standard_type"] == "IC50"]
```

# Handeling missing values

Let's drop the missing values for the standard_value which is going to be out target for the modeling later

```python
df1 = df[df.standard_value.notna()]
df1.head(3)
```

# Data preprocessing

* 🟢 Active (potent inhibitors) : IC50 ≤ 1 µM (1000 nM) , considered strong binders
* 🔴 Inactive : IC50 ≥ 10 µM (10,000 nM) , poor inhibition

```python
bioactivity_class = []
for i in df1.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append('active')
  else:
    bioactivity_class.append('intermmediate')
```

```python
columns = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
```

```python
df2 = df1[columns]
```

```python
dataset = pd.concat([df2, pd.Series(bioactivity_class, name='bioactivity_class')], axis=1)
```

```python
dataset
```

```python
dataset.to_csv("chembl_preprocessed.csv", index=False)
```
